```python
#
import zipfile

# def extract_zip_archive(zip_path, output_dir):
#     with zipfile.ZipFile(zip_path, 'r') as zip_ref:
#         zip_ref.extractall(output_dir)

def extract_zip_archive(zip_path, output_dir):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        for file_info in zip_ref.infolist():
            # Extract file
            zip_ref.extract(file_info, output_dir)

            # Get extracted file path
            extracted_file_path = os.path.join(output_dir, file_info.filename)

            # Generate UUID for new file name
            new_filename = str(uuid.uuid4())

            # Get file extension
            _, file_extension = os.path.splitext(extracted_file_path)

            # Rename file with UUID
            new_file_path = os.path.join(output_dir, new_filename + file_extension)
            os.rename(extracted_file_path, new_file_path)


# 입력 ZIP 파일 경로
zip_path = '/content/drive/MyDrive/github/Fake or Real 판별/Fake_or_Real/data/test_padding_224by224.zip'
output_dir = '/content/data/test/'
extract_zip_archive(zip_path, output_dir)

# 입력 ZIP 파일 경로
zip_path = '/content/drive/MyDrive/github/Fake or Real 판별/Fake_or_Real/data/train_padding_224by24.zip'
output_dir = '/content/data/train/'
extract_zip_archive(zip_path, output_dir)
```

```python
import os
import uuid
import shutil

# 중복 파일명을 피하기 위해 파일을 복사하고 새로운 파일명으로 저장
def save_non_duplicate_files(source_dir, destination_dir):
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    file_list = os.listdir(source_dir)
    for filename in file_list:
        source_path = os.path.join(source_dir, filename)
        new_filename = str(uuid.uuid4()) + os.path.splitext(filename)[-1]
        destination_path = os.path.join(destination_dir, new_filename)
        shutil.copyfile(source_path, destination_path)

source_dir_1 = '/content/data_crop/train/fake_images'
source_dir_2 = '/content/data_padding/train/fake_images'
destination_dir = '/content/data/train/fake_images'

save_non_duplicate_files(source_dir_1, destination_dir)
save_non_duplicate_files(source_dir_2, destination_dir)

source_dir_1 = '/content/data_crop/train/real_images'
source_dir_2 = '/content/data_padding/train/real_images'
destination_dir = '/content/data/train/real_images'

save_non_duplicate_files(source_dir_1, destination_dir)
save_non_duplicate_files(source_dir_2, destination_dir)

source_dir_1 = '/content/data_crop/test/images'
source_dir_2 = '/content/data_padding/test/images'
destination_dir = '/content/data/test/real_images'

save_non_duplicate_files(source_dir_1, destination_dir)
save_non_duplicate_files(source_dir_2, destination_dir)
```

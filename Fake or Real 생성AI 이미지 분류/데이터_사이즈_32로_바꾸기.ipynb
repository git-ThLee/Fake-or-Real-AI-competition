## 이미지 32x32 변환

```python
from PIL import Image
import os

def resize_images(input_dir, output_dir, size, output_format):
    # 출력 디렉토리가 없을 경우 생성
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 입력 디렉토리의 모든 이미지 파일 경로 가져오기
    image_paths = [os.path.join(input_dir, filename) for filename in os.listdir(input_dir) if filename.endswith(('.jpg', '.jpeg', '.png'))]

    # 이미지들을 32x32 크기로 변환하여 저장
    for image_path in image_paths:
        img = Image.open(image_path).convert('RGB')
        img = img.resize(size)

        # 이미지 파일명
        filename = os.path.basename(image_path)
        # 저장할 경로 및 파일 확장자
        output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + '.' + output_format.lower())
        img.save(output_path, format=output_format)
```

```python
# 이미지들을 32x32 크기로 변환하여 저장
# 3분40초
resize_images(
    '/content/data/train/fake_images', # input_dir
    '/content/data_32by32/train/fake_1', # output_dir
    (32, 32),       # 32 x 32
    'png' # output_format
)
```

```python
# 검토
import os

def count_files_in_directories(directory):
    # 디렉토리 안에 있는 서브 디렉토리 및 파일 리스트 가져오기
    subdirectories = [name for name in os.listdir(directory) if os.path.isdir(os.path.join(directory, name))]

    # 각 디렉토리의 파일 개수 출력
    for subdir in subdirectories:
        subdir_path = os.path.join(directory, subdir)
        file_count = len(os.listdir(subdir_path))
        print(f"{subdir}: {file_count} files")

# 디렉토리 경로 설정
directory_path = '/content/data_32by32/train'

# 각 디렉토리의 파일 개수 출력
count_files_in_directories(directory_path)
print('-'*50)
count_files_in_directories('/content/data_32by32/test')
```

```python
# 압축하는 코드
import shutil

def create_zip_archive(source_dir, output_path):
    shutil.make_archive(output_path, 'zip', source_dir)

# 원본 디렉토리 경로
source_dir = '/content/data_32by32'
# 출력 ZIP 파일 경로
output_path = '/content/data_32by32'

# 디렉토리를 ZIP 압축
create_zip_archive(source_dir, output_path)
```

```python
import os
from PIL import Image
import shutil

source_dir = "/content/AI_Images"

# 임시 디렉토리 생성
temp_dir = os.path.join(source_dir, "temp")
os.makedirs(temp_dir, exist_ok=True)

# 하위 디렉토리 탐색
for root, dirs, files in os.walk(source_dir):
    for filename in files:
        # 이미지 파일인지 확인
        if filename.endswith((".jpg", ".jpeg", ".gif", ".png")):
            filepath = os.path.join(root, filename)

            # 이미지 열기
            image = Image.open(filepath)

            # 파일 확장자 변경을 위한 임시 파일 경로 생성
            temp_filepath = os.path.join(temp_dir, filename)
            temp_filepath = os.path.splitext(temp_filepath)[0] + ".png"

            # 중복된 이름이 있을 경우 다른 이름으로 수정하여 저장
            count = 1
            while os.path.exists(temp_filepath):
                temp_filepath = os.path.join(temp_dir, f"{os.path.splitext(filename)[0]}_{count}.png")
                count += 1

            # 이미지를 PNG로 저장
            image.save(temp_filepath, "PNG")

            # 원본 파일 삭제
            os.remove(filepath)

# 이미지 파일을 옮길 디렉토리 생성
target_dir = "/content/AI_Images/fake/"
os.makedirs(target_dir, exist_ok=True)

# 임시 디렉토리의 파일을 목표 디렉토리로 이동
for filename in os.listdir(temp_dir):
    filepath = os.path.join(temp_dir, filename)
    shutil.move(filepath, target_dir)

# 임시 디렉토리 삭제
os.rmdir(temp_dir)

print("이미지 파일 이동 및 확장자 변경이 완료되었습니다.")
```

```python
import os
import shutil

def move_images(source_dir, target_dir):
    # 대상 디렉토리 생성
    os.makedirs(target_dir, exist_ok=True)

    # 이미지 파일을 대상 디렉토리로 이동
    for filename in os.listdir(source_dir):
        filepath = os.path.join(source_dir, filename)

        # 이미지 파일인지 확인
        if os.path.isfile(filepath) and filename.endswith((".jpg", ".jpeg", ".gif", ".bmp", ".png")):
            # 중복된 이름이 있을 경우 다른 이름으로 수정하여 이동
            count = 0
            new_filename = filename
            while os.path.exists(os.path.join(target_dir, new_filename)):
                count += 1
                name, ext = os.path.splitext(filename)
                new_filename = f"{name}_{count}{ext}"

            # 이미지 파일을 대상 디렉토리로 이동
            shutil.move(filepath, os.path.join(target_dir, new_filename))

    print("이미지 파일의 이동이 완료되었습니다.")

# 함수 호출
move_images("/content/AI_Images/fake", "/content/data/train/fake_images")
move_images("/content/AI_cat_dog_dataset/fake", "/content/data/train/fake_images")
move_images("/content/FACE_fake_or_real_300by300/fake", "/content/data/train/fake_images")

move_images("/content/FACE_fake_or_real_300by300/real", "/content/data/train/real_images")
move_images("/content/People_image_dataset/real", "/content/data/train/real_images")
```

## 1.Padding & Resize(224,224) & Save

```python
import os
from PIL import Image

def pad_images(input_dir, output_dir, target_size, background_color=(0, 0, 0)):
    # 디렉토리 생성
    os.makedirs(output_dir, exist_ok=True)

    # 경로에 있는 이미지 파일들을 패딩하여 저장
    for filename in os.listdir(input_dir):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # 이미지 파일 필터링
            image_path = os.path.join(input_dir, filename)
            image = Image.open(image_path)

            # 이미지를 패딩하여 크기 조정
            image.thumbnail(target_size, Image.ANTIALIAS)
            image_size = image.size

            # 패딩을 통해 최종적으로 target_size 크기로 조정
            if image_size[0] != target_size[0] or image_size[1] != target_size[1]:
                padding = (target_size[0] - image_size[0]) // 2, (target_size[1] - image_size[1]) // 2
                background = Image.new("RGB", target_size, background_color)
                background.paste(image, padding)
                image = background

            # 새로운 경로에 이미지 저장
            new_image_path = os.path.join(output_dir, filename)
            image.save(new_image_path)

    print("이미지 패딩 및 저장이 완료되었습니다.")
```

```python
# Fake train
input_dir = "/content/data/train/fake_images"
output_dir = "/content/padding/data/train/fake_images"
target_size = (224, 224)
pad_images(input_dir, output_dir, target_size)
```

## 2.Crop & Resize(224,224) & Save

- 가로, 세로 중 큰 쪽을 작은 쪽만큼 맞춰서 정사각형을 맞추고, 리사이즈

```python
import os
from PIL import Image

def resize_and_save_images(source_directory, target_directory, target_size):
    # 대상 디렉토리 생성
    os.makedirs(target_directory, exist_ok=True)

    # 디렉토리 내의 이미지 파일 목록 가져오기
    image_files = [file for file in os.listdir(source_directory) if file.lower().endswith((".png", ".jpg", ".jpeg", ".gif", ".bmp"))]

    # 이미지 파일마다 함수 적용
    for filename in image_files:
        source_path = os.path.join(source_directory, filename)
        target_path = os.path.join(target_directory, filename)

        # 이미지 열기
        image = Image.open(source_path)
        width, height = image.size

        # 가로, 세로 중 큰 쪽을 작은 쪽만큼 맞춰서 정사각형으로 맞추기
        if width > height:
            diff = (width - height) // 2
            image = image.crop((diff, 0, width - diff, height))
        else:
            diff = (height - width) // 2
            image = image.crop((0, diff, width, height - diff))

        # 리사이즈
        resized_image = image.resize(target_size, Image.ANTIALIAS)

        # 이미지 저장
        resized_image.save(target_path)
        #print(f"Image saved: {target_path}")qz
```

```python
import os
import math

base_dir = "/content/drive/MyDrive/github/Fake or Real 판별/Fake_or_Real"

# 용량을 GB 단위로 변환하는 함수
def convert_to_gb(size_in_bytes):
    size_in_gb = size_in_bytes / (1024 * 1024 * 1024)
    return round(size_in_gb, 2)

# 바로 하위 디렉토리 용량 출력
for entry in os.scandir(base_dir):
    if entry.is_dir():
        total_size = 0
        for dirpath, dirnames, filenames in os.walk(entry.path):
            for filename in filenames:
                file_path = os.path.join(dirpath, filename)
                total_size += os.path.getsize(file_path)
        dir_name = entry.name
        total_size_gb = convert_to_gb(total_size)
        print(f"{dir_name}: {total_size_gb} GB")
```
